```python
path = 'KML/KML//*.kml'   
files=glob.glob(path)  
count=0
zList=list()
for file in files:    
    f=open(file, 'r') 
    print(f.name[8:-4])
    count = 0
    zList = list()
    while 1:
        line = f.readline()
        if not line: break
        str1 = "<coordinates>"
        str2 = "</coordinates>"
    #     if (line.find(str1) != -1):
    # #         print(line[59:])
        if (line.find(str2) != -1):
            count += 1
            str3 = line[72:line.find(str2)]
            stry = str3.split(' ')
    #         print(stry)
            for elem in stry:
                str4 = elem.split(',')
    #             print(str4[0])
                zList.append([str4[0],str4[1],count]) 
    f.close() 
    with open(f.name[8:-4]+'.csv', 'w', newline='') as myfile:
     wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
     wr.writerows(zList)
```
